```python
import os

path='labeling/big_truck/x'

file_list = os.listdir(path)
file_list
```

```python
import json


merged_dict = {}

for file_name in file_list:
    file_path = os.path.join(path, file_name)
    with open(file_path, 'r', encoding='utf-8') as f:
        file_content = json.load(f)
        merged_dict[file_name] = file_content

# 결과를 하나의 JSON으로 저장
with open('merged_by_filename.json', 'w', encoding='utf-8') as out_f:
    json.dump(merged_dict, out_f, ensure_ascii=False, indent=2)
```

```python
box_dict = {}

for filename, items in data_final.items():
    # 각 항목의 'BOX' 값을 실수형 리스트로 파싱
    box_list = [list(map(float, item['BOX'].split(','))) for item in items]
    box_dict[filename] = box_list
box_dict
```

```python
import cv2
import os

# 박스 정보가 들어있는 딕셔너리
# 예: box_dict = {'filename.jpg': [[x1, y1, x2, y2], [...], ...]}
# 아래는 가정된 box_dict
# box_dict = {...}

# 원본 이미지 폴더 경로
input_dir = 'raw/big_truck_x'

# 출력 폴더 (없으면 생성)
output_dir = 'output'
os.makedirs(output_dir, exist_ok=True)

# 모든 이미지에 대해 처리
for filename, box_list in box_dict.items():

    safe_filename = filename.replace('\\', '/').split('/')[-1]
    input_path = os.path.join(input_dir, filename)
    output_path = os.path.join(output_dir, filename)

    # 이미지 읽기
    img = cv2.imread(input_path)
    if img is None:
        print(f"⚠️ 이미지 불러오기 실패: {input_path}")
        continue

    # 박스 그리기
    for box in box_list:
        x1, y1, x2, y2 = map(int, box)
        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 0, 255), 3)  # 빨간색, 두께 3

    # 이미지 저장
    cv2.imwrite(output_path, img)
    print(f"✅ 저장 완료: {output_path}")
```
